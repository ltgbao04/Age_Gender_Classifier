```python
import pandas as pd 
import numpy as np
import os
from os import path
import re
import tensorflow as tf 
import keras
from keras.preprocessing import image_dataset_from_directory
```

```python
file_names_p1 = [f for f in os.listdir('data\part1') if path.isfile(path.join('data\part1', f))]
file_names_p2 = [f for f in os.listdir('data\part2') if path.isfile(path.join('data\part2', f))]
file_names = file_names_p1 + file_names_p2
```

```python
age_labels = []
gender_labels = []

def get_labels(data):
    for i in range(len(data)):
        age, gender = re.split('_', data[i])[:2]
        age_labels.append(eval(age) if age else np.nan)
        gender_labels.append(eval(gender) if gender else np.nan)

get_labels(file_names)
```
